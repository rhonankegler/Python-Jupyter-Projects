A function that strictly takes the top card of the deck and shuffles it randomly into one of the 52 spots.

```python
def one_card(deck): #names the function 
    shuffled_deck = deck.copy() #the new deck will create a copy of the untouched deck
    top_card = shuffled_deck.pop(0) #removes the top card
    new_position = random.randint(0,51) #chooses randomly one of the 52 spots in the deck
    shuffled_deck.insert(new_position,top_card) #inserts the top card into the randomly chosen spot
    return shuffled_deck #saves the new deck
```

### Measuring Randomness
A function that will measure the randomness of a distribution, or in this case, the placement of the cards after a particular shuffle method is shown below. The function is computing the average distances between adjacent integers.

$$ [n_1,n_2,n_3,...n_k], $$

$$ \frac{1}{k} \sum_{i=1}^{k-1} {|n_{i+1}-n_i|} $$

For example if we have a list of the first 4 values:

```python
list = [1,2,3,4] #initializes a list of 4 values
random.shuffle(list) #randomizes the values in the list into a new order
print(list) #prints the new list
```

We have 3 total distances between the 4 adjacent integers. These are 3, 1, and 1 respectively.

$$ \frac{3+1+1}{3} = \frac{5}{3} $$

$ \frac{5}{3} $ would be the measure of randomness variable for the list.

## The Classic Riffle Method

This shuffling technique is arguably the most universally used and most common technique. The goal of this shuffle is to start by taking a deck and roughly dividing it in half. With two new stacks, one in each hand, the dealer begins flaring the cards starting with the bottom of each stack, and simulating an alternating stacking pattern until all the cards have formed a single new stack.

A perfect example of starting off the first step of the riffle.

```python
top_half = deck[:26] #takes the first half of the deck and creates a new list
bottom_half = deck[26:] #takes the second half of the deck and creates another list
print(top_half) #prints the first 26 cards as a list
print(bottom_half) #prints the last 26 cards as a list
```

It is unrealistic that everytime someone goes to perform a riffle they divide the deck perfectly even with 26 cards in each hand. To represent a more realistic action, we can impliment binomial distribution.

```python
from numpy.random import binomial #imports a module for binomial distribution
```

```python
deck = [i for i in range(1,53)] #starts with a list to represent an untouched deck
```

```python
def split_deck(deck): #names the function and references the original deck
    top_half = deck[:binomial(52,0.5)] #splits the top half of the deck based on a 0.5 binomial distribution
    bottom_half = [i for i in range(1,53) if i not in top_half] #the bottom half is the remainder of the original deck minus what was removed by the top half
    return [top_half,bottom_half] #saves 2 new lists of cards: top half and bottom half
```

```python
print(split_deck(deck)) #prints the list of the top half and bottom half
```

```python
def riffle_shuffle(deck): #names the riffle function and references the untouched deck
    x=[] #initializes an empty list which will be used to store the deck after it has been shuffled
    top_half,bottom_half = split_deck(deck) #two outcomes for splitting the deck
    for i in range(1,53): #the deck range
        selection = random.randint(0,1) #selects one of the two outcomes, top or bottom
        if selection == 0: #if the top half is selected
            last = top_half.pop(-1)  #removes the last card of the top half
        if selection == 1: #if the the bottom half is selected
            last = bottom_half.pop(-1) #removes the last card of the bottom half
        x.append(last) #adds the removed cards one by one to the bottom of the new list
        if len(top_half) == 0: #if the top half has transferred all of its cards
            x+=bottom_half #adds the rest of the bottom half to the new list
            break #ends the for loop
        if len(bottom_half) == 0: #if the bottom half has transferred all of its cards
            x+=top_half #adds the rest of the top half to the new list
            break #ends the for loop
    return x #saves the new shuffled deck
```

```python
print(riffle_shuffle(deck)) #prints the new deck as a list after one riffle shuffle
```

```python
def randomness(deck): #names the function that will measure randomness of a single shuffle
    total = 0 #will hold the value for the randomness measure
    for k in range(1,52): #the range for possible inputs
        total += abs(deck[k]-deck[k-1]) #adds the absolute value between each card value and its preceding card value
    total = total/51 #divides the total from the previous line by 51 since there is 51 distances between the 52 cards
    return total #saves the randomness measure
```

```python
deck =[i for i in range(1,53)] #establishes an unshuffled deck
y = [1] #the list will already start with 1 because the 1st orientation of cards has not changed since it unshuffled, meaning the randomness will be 1
for i in range(10): #shuffles the deck 10 times
    deck = riffle_shuffle(deck) #uses the riffle shuffle to update the deck
    y.append(randomness(deck)) #the y list is updated by adding the randomness after each of the 10 shuffles
```

## The Middle-Out Method

A method that will first divide the deck into roughly thirds. The middle third is then placed randomly on the top or the bottom of the deck.

```python
deck = [i for i in range(1,53)] #starts with a list to represent an untouched deck
```

```python
def middle_out(deck): #names the function
    x = binomial(52,0.33) #chooses a number to represent the top third of the deck based on a 0.33 binomial distribution of 52
    top_third = deck[:x] #takes cards up until the selected number based on previous line and separates it from the original deck
    deck = deck[x:] #the remainder of the deck is the orginal deck minus the top third taken off
    y = binomial(len(deck),0.5) #splits the other roughly two thirds in half based on a 0.5 binomial distribution
    middle_third = deck[:y] #middle third will be the first 0.5 of the distribution
    bottom_third = deck[y:] #bottom third will be the remainder
    if random.random() > 0.5: #the middle third will have a 50% to:
        return middle_third + top_third + bottom_third #if the random binomial is less than 0.5, place middle third on top
    else: #alternative outcome
        return top_third + bottom_third + middle_third #if the random binomial is greater than 0.5, place the middle third on the bottom 
```

```python
print(middle_out(deck)) #prints the deck after one shuffle of the middle out method
```

```python
deck =[i for i in range(1,53)] #establishes a new deck of 52 cards
y = [1] #randomness will start at 1 since it has not been shuffled yet
for i in range(10): #represents the 10 shuffles
    deck = middle_out(deck) #the middle out shuffle method will be applied to the deck
    y.append(randomness(deck)) #adds the randomness measure to a list after each shuffle
```

```python
x = [i for i in range(101)] #sets the range as the 100 shuffles
plt.plot(x,y) #uses the x axis as the number of trials and y axis as the measure of randomness
plt.title('Randomness of 100 Consecutive Middle-Out Shuffles') #title for the graph
plt.xlabel('Number of Shuffles') #title for the x-axis
plt.ylabel('Measure of Randomness') #title for the y-axis
plt.show() #hides unnecessary code that appears above graph
```

## The Top-to-Bottom Re-stack/Overhand Method

A method that begins by dividing the deck into a random number of sections. Then the deck is re-stacked by reversing the order of the sections.

```python
deck =[i for i in range(1,53)] #starts with a list to represent an untouched deck
```

```python
def overhand(deck,n): #names the function
    right_hand = deck #the right hand will have the whole deck to begin
    left_hand = [] #left hand with begin as an empty list
    for i in range(n): #range n will be the number of sections the deck is divided into
        x = binomial(52,1/(n-i)) #a binomial distribution based on the number of sections we choose to divide the deck into
        y = right_hand[:x] #the rest of the cards
        left_hand = y + left_hand #each section will from the righthand will be added to the left hand in reverse order
        right_hand = right_hand[x:] #the remainder of the right hand
    return left_hand #saves the new order of the deck after one shuffle
```

```python
print(overhand(deck,8)) #prints the new order of the deck if it seperated into 8 sections and shuffled
```

```python
deck =[i for i in range(1,53)] #establishes a new deck of 52 cards
y = [1] #randomness measure will start with 1 because the deck has not been shuffled to begin
for i in range(10): #there will be 10 shuffles
    deck = overhand(deck,8) #the deck will be shuffled using the overhand method and separated into 8 sections each time
    y.append(randomness(deck)) #the randomness measure after each shuffle will be added to a list
```

```python
deck =[i for i in range(1,53)] #establishes a new deck of 52 cards
y = [1] #randomness measure will start with 1 because the deck has not been shuffled to begin
for i in range(100): #there will be 100 shuffles
    deck = overhand(deck,8) #the deck will be shuffled using the overhand method and separated into 8 sections each time
    y.append(randomness(deck)) #the randomness measure after each shuffle will be added to a list
```

```python
x = [i for i in range(101)] #sets the range as the 100 shuffles
plt.plot(x,y) #uses the x axis as the number of trials and y axis as the measure of randomness
plt.title('Randomness of 100 Consecutive Overhand Shuffles')  #title for the graph
plt.xlabel('Number of Shuffles') #title for the x-axis
plt.ylabel('Measure of Randomness') #title for the y-axis
plt.show() #hides unnecessary code that appears above graph
```

## Observations
#### Method that reached most randomness
Out of the first 100 shuffles for each method, the overhand method ultimately achieved the highest randomness at about 20 while the riffle reached about 18 and the middle out reached roughly 17.5. 

#### Method that reached an acceptable level of randomness the fastest
An acceptable level of randomness can be interpreted in several ways. For the sake of shuffling cards in a timely manner we can have 5 shuffles as the ideal amount. We can take the average randomness after the 5th shuffle of the 3 methods and find the average to be roughly 7.8. The riffle shuffle achieves this randomness the quickest followed by the overhand and then the middle out.

#### Additional comments
We can see that both the overhand and middle out method increase in randomness over time but take longer to reach a higher level of randomness than the riffle does. Even though the riffle shuffle reaches a higher randomness quicker, it seems to average out at about 13 almost instantly and oscillates above and below this mean rather uniformly. Although you can ultimately achieve greater randomness using the overhand or middle-out, the riffle is much more efficient at achieving randomness because realistically not many people are shuffling a deck 10, 20 times before they deal the cards. The maximum randomness for the riffle occurs within the first few shuffles but this has high variance each time you run the 100 shuffles. The middle out will reach a maximum randomness typically between 60 and 80 shuffles and the overhand at about 80 to 100 shuffles.

## Conclusion
Overall the most efficient shuffle for a person playing a typical card game under time constraints would be the riffle shuffle. Over time the other two shuffles will become more random but will take far longer. This is based on the randomness measurement equation and the speed to which it reached.
